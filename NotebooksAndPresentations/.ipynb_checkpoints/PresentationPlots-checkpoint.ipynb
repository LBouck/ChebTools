```python
# code used from Ian Bell (NIST)
import sympy as sympy
def ChebyshevPolynomials(N):
 
    xx = np.linspace(-1,1,1000)
    x = sympy.symbols('x')
    plt.figure(figsize=(3.25,2.75))
    xlocs = {0:0.4, 1:0.55, 2:0.65, 3:0.75, 4:0.8}
    for n in range(N):
        f_i = sympy.lambdify(x, sympy.chebyshevt_poly(n, x=x), ("numpy"))
        line, = plt.plot(xx, [f_i(_x) for _x in xx])
        if n%2 == 0:
            kmax = n//2
        else:
            kmax = 1+(n-1)//2
        plt.text(xlocs[n], f_i(xlocs[n]),'$T_{{{n:d}}}$'.format(n=n), bbox=dict(facecolor='w',edgecolor='none',pad=0), color=line.get_color(),ha='center',va='center')
        if n > 0:
            for k in range(1, 2+int((n-1)//2)):
                xk = np.cos(((2*k-1)*np.pi)/(2*n))
                print(n, kmax, k, xk, ((2*k-1)*np.pi)/(2*n))
                plt.plot(xk,0,'o',color=line.get_color())
                plt.plot(-xk,0,'o',color=line.get_color())
            if n%2==0:
                plt.plot(0,0,'o',color=line.get_color())
 
    plt.xlabel('$\chi$')
    plt.ylabel('$y$')
    plt.ylim(-1.1, 1.1)
    plt.tight_layout(pad=0.3)
    plt.savefig('cheb_poly_plot.pdf',transparent=True)
    plt.close()
ChebyshevPolynomials(5)
```

```python
#interpolation ex plot
def f(x):
    return np.exp(x)

cheb2 = ct.generate_Chebyshev_expansion(2,f,-1,1)
xvals = np.linspace(-1.5,1.5)
node_vals = cheb2.get_nodes_n11()
plt.figure(figsize=(6,4))
plt.plot(xvals,f(xvals),'k', linewidth=2, label=r"$e^x$")
plt.plot(xvals,cheb2.y(xvals),'b',linewidth=2,label=r"$p_2(x)$")
plt.plot(node_vals,f(node_vals),'ro',label=r"Interpolation Points")
plt.legend(bbox_to_anchor=(.52,1))
plt.title(r"Example of Quadratic Interpolation of $e^x$")
plt.xlabel("x")
plt.ylabel("y")
plt.savefig("interpolation_ex.pdf")
```

```python
# cheb extrema nodes plot

xvals = np.linspace(-1,1,100)
cheb2 = ct.generate_Chebyshev_expansion(15,f,-1,1)
node_vals = cheb2.get_nodes_n11()
plt.figure(figsize=(6,4))
circle = np.sqrt(1-xvals**2)
circle[0] = 0
circle[-1] = 0
node_circle = np.sqrt(1-node_vals**2)
node_circle[0] = 0
node_circle[-1] = 0
plt.plot(xvals,circle,'k', linewidth=2)
plt.plot(node_vals,node_circle,'ro')
plt.plot(node_vals,np.zeros_like(node_vals),'bo')
for n in node_vals:
    plt.plot(np.ones(10)*n,np.linspace(np.sqrt(1-n**2),0,10),'k--')
plt.title(r"Chebyshev Extrema Nodes with $N=15$")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([-1.01,1.01])
plt.ylim([-.01,1.01])
plt.savefig("extrema_nodes.pdf")
```

```python
def f(x):
    return 1/(1+25*x**2)
mpl.rcParams['axes.titlesize'] = 12
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12

xvals = np.linspace(-1,1,1000)
N = 200
error = np.zeros(N)
for n in range(N):
    cheb = ct.generate_Chebyshev_expansion(n+2,f,-1,1)
    error[n] = np.max(np.abs(f(xvals)-cheb.y(xvals)))
nvals = np.arange(2,N+2)
plt.figure(figsize=(5,3.5))
plt.loglog(nvals,error,'b',linewidth=2,label=r"$|| f-p_n ||$")
plt.loglog(np.arange(1,N+2)[2::2],np.abs(cheb.coef()[2::2]),'r',linewidth=2,label=r"$|a_n|$")
plt.grid()
plt.legend(bbox_to_anchor=(.5,.5))
plt.title("Error of Chebyshev Interpolation of Runge Function")
plt.xlabel(r"$n$")
plt.savefig("runge_cheb_error.pdf")
```

```python
def f(x):
    if x<=0:
        return 0*x
    else:
        return x**3
def vec_f(xs):
    ans = np.zeros(len(xs))
    for i, x in enumerate(xs):
        ans[i] = f(x)
    return ans




N = 100
error_array2 = np.zeros(N)

xvals1 = np.linspace(-1,1,100)
for i in range(1,N+1):
    error_array2[i-1] = np.max(np.abs(vec_f(xvals1)-ct.generate_Chebyshev_expansion(i,f,-1,1).y(xvals1)))
```

```python
def g(x):
    if x<=0:
        return 0*x
    else:
        return x**5
def vec_g(xs):
    ans = np.zeros(len(xs))
    for i, x in enumerate(xs):
        ans[i] = g(x)
    return ans

N = 100
error_array4 = np.zeros(N)
for i in range(1,N+1):
    error_array4[i-1] = np.max(np.abs(vec_g(xvals1)-ct.generate_Chebyshev_expansion(i,g,-1,1).y(xvals1)))
```

```python
def h(x):
    if x<=0:
        return 0*x
    else:
        return x**7
def vec_h(xs):
    ans = np.zeros(len(xs))
    for i, x in enumerate(xs):
        ans[i] = h(x)
    return ans

N = 100
error_array6 = np.zeros(N)
for i in range(1,N+1):
    error_array6[i-1] = np.max(np.abs(vec_h(xvals1)-ct.generate_Chebyshev_expansion(i,h,-1,1).y(xvals1)))
```

```python
def f(x):
    return np.exp(x)

xvals = np.linspace(-1,1,1000)
error_arrayinf = np.zeros(N)
for i in range(1,N+1):
    error_arrayinf[i-1] = np.max(np.abs(f(xvals)-ct.generate_Chebyshev_expansion(i,f,-1,1).y(xvals)))
```

```python
plt.figure()
plt.loglog(np.arange(1,N+1),error_array2,'k',label=r"$\nu = 3$")
plt.loglog(np.arange(1,N+1),error_array4,'b',label=r"$\nu = 5$")
plt.loglog(np.arange(1,N+1),error_array6,'c',label=r"$\nu = 7$")
plt.loglog(np.arange(1,N+1),error_arrayinf,'r',label=r"Analytic")
plt.xlabel(r"Degree of Chebyshev Interpolation")
plt.ylabel(r"Max Abs Error")
plt.title(r"Interpolation Error of Chebyshev Polynomials")
plt.grid()
plt.legend(bbox_to_anchor=(.38,.45))
plt.savefig("cheb_convergence_plots.pdf")
```

```python
from numpy.linalg import LinAlgError
def newton(f,g,dfdx,dfdy,dgdx,dgdy,x0):
    tol = 1e-14
    err = 1
    x = x0
    z = np.zeros(2)
    jacobian = np.zeros((2,2))
    count = 0
    while err>tol:
        if count>10:
            break
        value = np.array([f(x),g(x)])
        jacobian = np.array([[dfdx(x),dfdy(x)],[dgdx(x),dgdy(x)]])
        try:
            z = np.linalg.solve(jacobian,value)
        except LinAlgError:
            if np.max(np.abs(value))<tol:
                return x
        x = x - z
        err = np.max(np.abs(z))
        count += 1
    return x
def f(x):
    return x[0]**2-x[1]**2
def dfdx(x):
    return x[0]*2
def dfdy(x):
    return -2*x[1]
def g(x):
    return np.sin(np.pi*np.sqrt(x[0]**2+x[1]**2))
def dgdx(x):
    return np.pi*x[0]*np.cos(np.pi*np.sqrt(x[0]**2+x[1]**2))/np.sqrt(x[0]**2+x[1]**2)
def dgdy(x):
    return np.pi*x[1]*np.cos(np.pi*np.sqrt(x[0]**2+x[1]**2))/np.sqrt(x[0]**2+x[1]**2)
.001
print(newton(f,g,dfdx,dfdy,dgdx,dgdy,np.array([0.001,.1])))
```

```python
X = np.linspace(-1, 1, 201)
Y = np.linspace(-1, 1,201)
list1 = []
root1 = np.array([np.sqrt(2)/2,np.sqrt(2)/2])
list2 = []
root2 = np.array([-np.sqrt(2)/2,np.sqrt(2)/2])
list3 = []
root3 = np.array([-np.sqrt(2)/2,-np.sqrt(2)/2])
list4 = []
root4 = np.array([np.sqrt(2)/2,-np.sqrt(2)/2])
list0 = []
root0 = np.array([0,0])
list5 = []
tol = 1e-10
for x in X:
    for y in Y:
        x0 = np.array([x,y])
        root = newton(f,g,dfdx,dfdy,dgdx,dgdy,x0)
        if np.linalg.norm(root-root0)<tol:
            list0.append(x0)
        elif np.linalg.norm(root-root1)<tol:
            list1.append(x0)
        elif np.linalg.norm(root-root2)<tol:
            list2.append(x0) 
        elif np.linalg.norm(root-root3)<tol:
            list3.append(x0)
        elif np.linalg.norm(root-root4)<tol:
            list4.append(x0)
        else:
            list5.append(x0)
arr0 = np.vstack(list0)
arr1 = np.vstack(list1)
arr2 = np.vstack(list2)
arr3 = np.vstack(list3)
arr4 = np.vstack(list4)
arr5 = np.vstack(list5)
```
